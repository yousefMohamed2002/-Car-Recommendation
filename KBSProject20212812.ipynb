```python
# import gradio as gr

# # Expert system function
# def recommend_car(gender, age, job, income, usage, family_size, fuel_type, driving_experience, weather, eco_conscious, safety_priority, parking_space, driving_style):
#     """Predicts the best car for a person based on multiple attributes."""

#     # Base recommendation based on income
#     if income > 80000:
#         car = "Tesla Model S (Luxury Electric)"
#     elif income > 50000:
#         car = "BMW 3 Series (Premium Sedan)"
#     elif income > 30000:
#         car = "Toyota Camry (Reliable Mid-size Sedan)"
#     else:
#         car = "Honda Civic (Budget-Friendly Compact Car)"

#     # Adjust based on usage
#     if usage == "Daily City Commute":
#         car = "Toyota Prius (Fuel Efficient Hybrid)"
#     elif usage == "Long Trips":
#         car = "Ford Explorer (Spacious SUV)"
#     elif usage == "Business":
#         car = "Mercedes-Benz E-Class (Executive Sedan)"

#     # Adjust based on family size
#     if family_size >= 4:
#         car = "Honda Odyssey (Spacious Minivan)"

#     # Adjust based on fuel preference
#     if fuel_type == "Electric":
#         car = "Tesla Model 3 (Electric Car)"
#     elif fuel_type == "Diesel":
#         car = "Ford F-150 (Powerful Diesel Truck)"

#     # Adjust based on driving experience
#     if driving_experience < 1:
#         car = "Toyota Yaris (Beginner-Friendly Compact Car)"

#     # 1ï¸âƒ£ Adjust for Weather & Road Conditions
#     if weather == "Snowy" or weather == "Rainy":
#         car = "Subaru Outback (AWD for Tough Conditions)"
#     elif weather == "Off-Road/Rough Terrain":
#         car = "Jeep Wrangler (Off-Road Capable SUV)"

#     # 2ï¸âƒ£ Environmental Concerns
#     if eco_conscious:
#         car = "Hyundai Ioniq 5 (Eco-Friendly Electric Car)"

#     # 3ï¸âƒ£ Safety Priorities
#     if safety_priority:
#         car = "Volvo XC90 (Top Safety Features & Family-Friendly)"

#     # 4ï¸âƒ£ Parking & Space Considerations
#     if not parking_space:
#         car = "Mini Cooper (Compact & Easy to Park)"

#     # 5ï¸âƒ£ Driving Style & Personality
#     if driving_style == "Sporty":
#         car = "Mazda MX-5 Miata (Fun Sports Car)"
#     elif driving_style == "Luxury":
#         car = "Lexus RX (High-End Comfort & Features)"

#     return f"ðŸš— Recommended Car: {car}"

# # Gradio UI
# iface = gr.Interface(
#     fn=recommend_car,
#     inputs=[
#         gr.Dropdown(["Male", "Female"], label="Gender"),
#         gr.Slider(18, 70, label="Age"),
#         gr.Textbox(label="Job"),
#         gr.Slider(10000, 100000, step=5000, label="Income"),
#         gr.Dropdown(["Daily City Commute", "Weekends", "Long Trips", "Business"], label="Usage"),
#         gr.Slider(1, 7, label="Family Size"),
#         gr.Dropdown(["Petrol", "Diesel", "Electric", "Hybrid"], label="Preferred Fuel Type"),
#         gr.Slider(0, 20, label="Years of Driving Experience"),
#         gr.Dropdown(["Sunny", "Snowy", "Rainy", "Off-Road/Rough Terrain"], label="Weather Conditions"),
#         gr.Checkbox(label="Are you eco-conscious?"),
#         gr.Checkbox(label="Do you prioritize safety?"),
#         gr.Checkbox(label="Do you have limited parking space?"),
#         gr.Dropdown(["Normal", "Sporty", "Luxury"], label="Driving Style"),
#     ],
#     outputs="text",
#     title="ðŸš˜ Expert System: Car Recommendation",
#     description="Answer the questions below to find the best car for you!",
# )

# iface.launch()
!pip install --upgrade frozendict

from experta import *
import gradio as gr

# Define the facts (attributes)
class CarPreferences(Fact):
    pass  # Fields will be automatically created based on initialization

# Define the expert system
class CarRecommender(KnowledgeEngine):
    def __init__(self):
        super().__init__()
        self.recommendation = ""

    @Rule(CarPreferences(income=P(lambda x: x > 80000)))
    def high_income(self):
        self.recommendation = "Tesla Model S (Luxury Electric)"

    @Rule(CarPreferences(income=P(lambda x: 50000 < x <= 80000)))
    def mid_high_income(self):
        self.recommendation = "BMW 3 Series (Premium Sedan)"

    @Rule(CarPreferences(income=P(lambda x: 30000 < x <= 50000)))
    def mid_income(self):
        self.recommendation = "Toyota Camry (Reliable Mid-size Sedan)"

    @Rule(CarPreferences(income=P(lambda x: x <= 30000)))
    def low_income(self):
        self.recommendation = "Honda Civic (Budget-Friendly Compact Car)"

    @Rule(CarPreferences(usage="Daily City Commute"))
    def city_commute(self):
        self.recommendation = "Toyota Prius (Fuel Efficient Hybrid)"

    @Rule(CarPreferences(usage="Long Trips"))
    def long_trips(self):
        self.recommendation = "Ford Explorer (Spacious SUV)"

    @Rule(CarPreferences(usage="Business"))
    def business(self):
        self.recommendation = "Mercedes-Benz E-Class (Executive Sedan)"

    @Rule(CarPreferences(family_size=P(lambda x: x >= 4)))
    def large_family(self):
        self.recommendation = "Honda Odyssey (Spacious Minivan)"

    @Rule(CarPreferences(fuel_type="Electric"))
    def electric(self):
        self.recommendation = "Tesla Model 3 (Electric Car)"

    @Rule(CarPreferences(fuel_type="Diesel"))
    def diesel(self):
        self.recommendation = "Ford F-150 (Powerful Diesel Truck)"

    @Rule(CarPreferences(driving_experience=P(lambda x: x < 1)))
    def beginner_driver(self):
        self.recommendation = "Toyota Yaris (Beginner-Friendly Compact Car)"

    @Rule(CarPreferences(weather=L("Snowy") | L("Rainy")))
    def bad_weather(self):
        self.recommendation = "Subaru Outback (AWD for Tough Conditions)"

    @Rule(CarPreferences(weather="Off-Road/Rough Terrain"))
    def off_road(self):
        self.recommendation = "Jeep Wrangler (Off-Road Capable SUV)"

    @Rule(CarPreferences(eco_conscious=True))
    def eco_friendly(self):
        self.recommendation = "Hyundai Ioniq 5 (Eco-Friendly Electric Car)"

    @Rule(CarPreferences(safety_priority=True))
    def safety(self):
        self.recommendation = "Volvo XC90 (Top Safety Features & Family-Friendly)"

    @Rule(CarPreferences(parking_space=False))
    def small_parking(self):
        self.recommendation = "Mini Cooper (Compact & Easy to Park)"

    @Rule(CarPreferences(driving_style="Sporty"))
    def sporty(self):
        self.recommendation = "Mazda MX-5 Miata (Fun Sports Car)"

    @Rule(CarPreferences(driving_style="Luxury"))
    def luxury(self):
        self.recommendation = "Lexus RX (High-End Comfort & Features)"

# Wrapper to run engine
def recommend_car_experta(gender, age, job, income, usage, family_size, fuel_type, driving_experience, weather, eco_conscious, safety_priority, parking_space, driving_style):
    engine = CarRecommender()
    engine.reset()
    engine.declare(CarPreferences(
        gender=gender,
        age=age,
        job=job,
        income=income,
        usage=usage,
        family_size=family_size,
        fuel_type=fuel_type,
        driving_experience=driving_experience,
        weather=weather,
        eco_conscious=eco_conscious,
        safety_priority=safety_priority,
        parking_space=parking_space,
        driving_style=driving_style
    ))
    engine.run()
    return f"ðŸš— Recommended Car: {engine.recommendation if engine.recommendation else 'No recommendation found'}"

# Gradio UI
iface = gr.Interface(
    fn=recommend_car_experta,
    inputs=[
        gr.Dropdown(["Male", "Female"], label="Gender"),
        gr.Slider(18, 70, label="Age"),
        gr.Textbox(label="Job"),
        gr.Slider(10000, 100000, step=5000, label="Income"),
        gr.Dropdown(["Daily City Commute", "Weekends", "Long Trips", "Business"], label="Usage"),
        gr.Slider(1, 7, label="Family Size"),
        gr.Dropdown(["Petrol", "Diesel", "Electric", "Hybrid"], label="Preferred Fuel Type"),
        gr.Slider(0, 20, label="Years of Driving Experience"),
        gr.Dropdown(["Sunny", "Snowy", "Rainy", "Off-Road/Rough Terrain"], label="Weather Conditions"),
        gr.Checkbox(label="Are you eco-conscious?"),
        gr.Checkbox(label="Do you prioritize safety?"),
        gr.Checkbox(label="Do you have limited parking space?"),
        gr.Dropdown(["Normal", "Sporty", "Luxury"], label="Driving Style"),
    ],
    outputs="text",
    title="ðŸš˜ Expert System: Car Recommendation (Powered by Experta)",
    description="Answer the questions below to find the best car for you!"
)

iface.launch()





```
